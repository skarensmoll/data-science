### cleaning data with orders that have more than 45 to collect payment

```python
import numpy as np
ordersMinMaxDate = data[['ORDER_ID', 'PAYMENT_DATE']].groupby(['ORDER_ID']).agg(min_date=('PAYMENT_DATE', np.min), max_date=('PAYMENT_DATE', np.max))
ordersMinMaxDate['DiffDays'] = (ordersMinMaxDate['max_date'] - ordersMinMaxDate['min_date']).dt.days
ordersMinMaxDate.reset_index(level=0, inplace=True)
```

```python
##Checking the diffs on days were calculated properly
specOrder = ordersMinMaxDate[ordersMinMaxDate['ORDER_ID'] == 214167721]
```

```python
ordersMoreFortyFiveDays = ordersMinMaxDate[ordersMinMaxDate.DiffDays > 45].ORDER_ID.values
```

```python
data.drop(data.loc[data['ORDER_ID'].isin(ordersMoreFortyFiveDays)].index, inplace=True)
```

### Deleting not paid orders

```python
pivotData = pivotData.loc[pivotData.PAID_FULL[~pivotData.PAID_FULL.isna()].index]
```

```python
paid_full_nan = pivotData.PAID_FULL[pivotData.PAID_FULL.isna()]
paid_full_nan
```

```python
def classify_payment(p):
    if p <= 15:
        ### On Time
        return 1
    elif p <= 30:
        ### Acceptable Late
        return 2
    else:
        ### Late
        return 3
```

```python
pivotData['TARGET'] = pivotData['PAID_FULL'].apply(lambda p : classify_payment(p))
```

```python
from matplotlib import pyplot as plt

plt.hist(pivotData['PAID_FULL'].sort_values(ascending=False))
```

### Split Data Set

```python
y = pivotData.pop('TARGET')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(pivotData, y, test_size=0.2, stratify=y, random_state=0)

print(X_train.shape,X_test.shape )
```

```python
X_train = pd.concat([X_train, y_train], axis=1)
```

```python
X_test = pd.concat([X_test, y_test], axis=1)
```
