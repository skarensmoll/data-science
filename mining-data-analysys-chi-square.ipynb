### Discretization using Decision Tree

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
y_train = train_imputed.pop('class')
```

```python
X_train = train_imputed
```

```python
kbins = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
data_trans = pd.DataFrame(kbins.fit_transform(X_train), columns= X_train.columns)
```

```python
data_trans
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import SelectPercentile
```

```python
## First Array represents chi-square values and second represents p-values
chi_scores = chi2(data_trans, y_train)
```

```python
p_values = pd.Series(chi_scores[1],index = data_trans.columns)
p_values.sort_values(ascending = False , inplace = True)
```

```python
p_values.plot.bar()
```
